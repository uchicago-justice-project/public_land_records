```python
import pandas as pd
```

```python
df = pd.read_csv("../../data/cook_county_cleaned2.csv", index_col=0)
```

```python
df.head()
```

```python
df.loc[df['Duplicate'] != True, 'Duplicate'] = False
```

```python
count = 0
for idx, row in df.iterrows():
    count = count + 1
    if count % 1000 == 0:
        print(count)
        
    if pd.notnull(row['Cleaned Name']):
        continue
    
    matches = [x for x in map(int, row['Name Matches'].split(";"))]
    if len(matches) < 2:
        df.loc[idx, "Cleaned Name"] = row['Purchaser']
        continue
    
    
    name = df.loc[matches[0], 'Purchaser']
    mismatched = False
    for i in matches[1:]:
        if df.loc[i, 'Duplicate'] == True:
            continue
        if df.loc[i, 'Purchaser'] != name:
            mismatched = True
            break
    
    if mismatched:
        continue
    
    df.loc[matches, 'Cleaned Name'] = name
```
